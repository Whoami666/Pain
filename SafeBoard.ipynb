```python
from bs4 import BeautifulSoup
import requests
import re

with open("/tlds-from-a-z.html") as fp:
    soup = BeautifulSoup(fp, "html.parser")

name = "feature-list"
elems = soup.find_all(name)
souper = str(soup.find_all("a", class_ ="smaller80"))
print(souper)

all_small_tlds = re.findall(r">\.[\w|\.]*", souper)
all_small_tlds = [x.strip(">.") for x in all_small_tlds]
all_small_tlds[:10]
```

```python
final_tlds_list = set(all_small_tlds + all_tlds)
```

```python
s = 'aboba abobovich'
finder = 'aboba'
len(re.findall(finder, s))
```

```python
def get_2ld_part1(domain: str, tld_set):
    parts = domain.split(".")
    if parts[-1] in tld_set:
        if ".".join(parts[-2:]) in tld_set:
            return parts[-3:][0] #, 
        else:
            return parts[-2:][0] #, 
    else:
        return parts[-2:][0] #, 

def get_2ld_part2(domain: str, tld_set):
    parts = domain.split(".")
    if parts[-1] in tld_set:
        if ".".join(parts[-2:]) in tld_set:
            return parts[-3:][-1] #, 
        else:
            return parts[-2:][-1] #, 
    else:
        return parts[-2:][-1] #, 
```

```python
df_good["second_domen_part1"] = df_good["domen"].apply(get_2ld_part1, args=[final_tlds_list])
df_good["second_domen_tdld"] = df_good["domen"].apply(get_2ld_part2, args=[final_tlds_list])
```

```python
df_good["joined_thing"] = df_good["second_domen_part1"] + '.' + df_good["second_domen_tdld"]
df_good.head(20)
```
